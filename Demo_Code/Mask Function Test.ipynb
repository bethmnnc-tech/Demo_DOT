#Demo of Column Masks

### Raw employee table

```python
%sql
Select birth_date, ssn, salary, * from main.dot_gold.dot_employee
```

### Create function for Date of Birth Mask

```python
%sql
CREATE OR REPLACE FUNCTION main.dot_gold.DOB_Mask (dt DATE)
RETURN cast('1900-01' as DATE);
```

###Alter DOB column to set the mask

```python
%sql
ALTER TABLE main.dot_gold.dot_employee
ALTER COLUMN birth_date SET MASK main.dot_gold.DOB_Mask;
```

### Create SSN and Salary Masks

```python
%sql
CREATE OR REPLACE FUNCTION main.dot_gold.ssn_mask (ssn STRING)
RETURN '***-**-****';
```

```python
%sql
CREATE OR REPLACE FUNCTION main.dot_gold.Salary_Mask (salary STRING)
RETURN 'REDACTED';
```

Add SSN and Salary Masks to employee table

```python
%sql
ALTER TABLE main.dot_gold.dot_employee
ALTER COLUMN salary SET MASK main.dot_gold.Salary_Mask;
```

```python
%sql
ALTER TABLE main.dot_gold.dot_employee
ALTER COLUMN ssn SET MASK main.dot_gold.SSN_Mask;
```
